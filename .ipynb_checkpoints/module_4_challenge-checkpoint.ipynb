```python
# Import necessary libraries
import pandas as pd
import numpy as np
from pathlib import Path
%matplotlib inline
import seaborn as sns
```

```python
#Set file paths
whale_path = Path(r"C:\Users\User\Documents\Manan Wala\Fintech bootcamp\Homework\Homework_3\Resources\whale_returns.csv")
algo_path = Path(r"C:\Users\User\Documents\Manan Wala\Fintech bootcamp\Homework\Homework_3\Resources\algo_returns.csv")
sp_tsx_path = Path(r"C:\Users\User\Documents\Manan Wala\Fintech bootcamp\Homework\Homework_3\Resources\sp_tsx_history.csv")
```

```python
# Read CSV files into individual DataFrames and set the 'date' column as a datetime index
whale_df = pd.read_csv(whale_path, index_col="Date", parse_dates=True, infer_datetime_format=True)
algo_df = pd.read_csv(algo_path, index_col="Date", parse_dates=True, infer_datetime_format=True)
sp_tsx_df = pd.read_csv(sp_tsx_path, index_col="Date", parse_dates=True, infer_datetime_format=True)
```

```python
# Sort the date index in ascending order
whale_df = whale_df.sort_index()
algo_df = algo_df.sort_index()
sp_tsx_df = sp_tsx_df.sort_index()
```

```python
# Read sample data in Whale Returns file
whale_df.head()
```

```python
# Identify null records in Whale Returns file
whale_df.isnull().mean()*100
```

```python
# Drop null records in Whale Returns file
whale_df = whale_df.dropna().copy()
```

```python
# Validate to check if null records have been dropped
whale_df.isnull().sum()
```

```python
# Read sample data in Algo Returns file
algo_df.head()
```

```python
# Identify null records in Algo Returns file
algo_df.isnull().mean()*100
```

```python
# Drop null records in Algo Returns file
algo_df = algo_df.dropna().copy()
```

```python
# Validate to check if null records have been dropped
algo_df.isnull().sum()
```

```python
# Read sample data in S&P TSX file
sp_tsx_df.head()
```

```python
# Identify null records in S&P TSX file
sp_tsx_df.isnull().mean()*100
```

```python
# Drop null records, if any, in S&P TSX file
sp_tsx_df = sp_tsx_df.dropna().copy()
```

```python
# Validate to check if null records have been dropped
sp_tsx_df.isnull().sum()
```

```python
# Clean S&P TSX file "Close" series by replacing $
sp_tsx_df["Close"] = sp_tsx_df["Close"].str.replace("$", "")
```

```python
# Verify S&P TSX file for "$"
sp_tsx_df.head()
```

```python
# Clean S&P TSX file "Close" series by replacing ,
sp_tsx_df["Close"] = sp_tsx_df["Close"].str.replace(",", "")
```

```python
# Confirm data type of "Close" series in S&P TSX file
sp_tsx_df["Close"].dtype
```

```python
# Cast "Close" series as float and then validate using dtype
sp_tsx_df["Close"] = sp_tsx_df["Close"].astype('float')
sp_tsx_df["Close"].dtype
```

```python
# Verify S&P TSX file for "$" and ","
sp_tsx_df.head()
```

```python
# Calculate daily returns for S&P TSX file
sp_tsx_daily_return = sp_tsx_df.pct_change()
sp_tsx_daily_return.head()
```

```python
# Concatenate all three files using columns axis and inner join
combined_returns = pd.concat([whale_df, algo_df, sp_tsx_daily_return], axis="columns", join="inner")
combined_returns.head()
```

```python
# Renaming last column to S&P TSX using rename function and set the columns parameter to a dictionary of new column names
combined_returns = combined_returns.rename(columns={
    "Close":"S&P TSX"})
combined_returns.head()
```

```python
# Plot daily returns for all portfolios
combined_returns.plot(figsize=(20,10))
```

```python
# Calculate cumulative return for all portfolios
cumulative_return = (1+combined_returns).cumprod()-1
cumulative_return.head()
```

```python
# Plot cumulative return for all portfolios
cumulative_return.plot(figsize=(20,10))
```

```python
# Does any portfolio outperform the S&P TSX 60?
# Answer: As can be seen from the above graph, the following portfolios outperform the S&P TSX 60:
        # Soros Fund Management LLC
        # Berkshire Hathaway Inc
        # Algo 1
        # Algo 2
```

```python
# Create a box plot for all portfolios
combined_returns.plot.box(figsize=(20,10))
```

```python
# Calculate the standard deviation for each portfolio & sort in ascending order
combined_returns_sd = combined_returns.std().sort_values(ascending = True)
print(f"Standard deviation for each portfolio is as under (in ascending order): \n{combined_returns_sd}")
```

```python
# Determine which portfolios are riskier than the S&P TSX 60
# Answer: As can be seen from the above standard deviation results for each portfolio, the following portfolios are riskier than the S&P TSX 60:
        # Algo 1
        # Soros Fund Management LLC
        # Algo 2
        # Tiger Global Management LLC
        # Berkshire Hathaway Inc
```

```python
# Calculate the annualised standard deviation for each portfolio
annualised_sd = combined_returns_sd*np.sqrt(252)
print(f"Annualised standard deviation for each portfolio is as under:\n{annualised_sd}")
```

```python
# Calculate and plot the rolling standard deviation for all portfolios, using a 21-day window
combined_returns.rolling(window=21).std().plot(figsize=(20,10))
```

```python
# Calculate and plot the correlation between each stock to determine which portfolios mimic the S&P TSX 60
combined_returns_correlation = combined_returns.corr()
sns.heatmap(combined_returns_correlation, vmin = -1, vmax = 1)
```

```python
# Display the correlation matrix of all portfolios
combined_returns_correlation
```

```python
# Which portfolios mimic the S&P TSX 60?
# Answer: Considering the above heatmap and correlation results, Algo 2 portfolio's correlation to S&P TSX 60 is the highest, Algo 2 closest mimics the S&P TSX 60.
```

```python
# Choose one portfolio, then calculate and plot the 60-day rolling beta between that portfolio and the S&P TSX 60.
# Chosen portfolio: Algo 1
```

```python
# Calculate and display the covariance of Algo 1 portfolio
algo_1_covariance = combined_returns["Algo 1"].cov(combined_returns["S&P TSX"])
algo_1_covariance
```

```python
# Calculate and display the variance of S&P TSX 60
index_variance = combined_returns["S&P TSX"].var()
index_variance
```

```python
# Calculate and display the beta for Algo 1 portfolio
algo_1_beta = algo_1_covariance/index_variance
algo_1_beta
```

```python
# Calculate and display the 60-day rolling covariance of Algo 1 portfolio
algo_1_rolling_covariance = combined_returns["Algo 1"].rolling(window=60).cov(combined_returns["S&P TSX"])
algo_1_rolling_covariance
```

```python
# Calculate and display the 60-day rolling variance of S&P TSX 60
index_variance_rolling = combined_returns["S&P TSX"].rolling(window=60).var()
index_variance_rolling
```

```python
# Calculate and plot the 60-day rolling beta between Algo 1 portfolio and the S&P TSX 60
rolling_algo1_beta = algo_1_rolling_covariance/index_variance_rolling
ax = index_variance_rolling.plot(figsize=(20,10), title="60-day rolling beta between Algo 1 and the S&P TSX 60")
rolling_algo1_beta.plot(ax=ax)
ax.legend(["S&P TSX 60", "Algo 1"])
```
